```python
def extraer_tabla_oscars():
    url_oscars = "https://es.wikipedia.org/wiki/Premios_%C3%93scar"

    # Extraer la tabla de los Premios Óscar y almacenarla en un DataFrame
    res_oscars = requests.get(url_oscars)

    # Verificar si la solicitud fue exitosa
    if res_oscars.status_code == 200:
        # Crear el objeto BeautifulSoup para acceder al contenido HTML
        sopa_oscars = BeautifulSoup(res_oscars.content, 'html.parser')

        # Encontrar todas las tablas con la clase 'wikitable'
        tablas = sopa_oscars.find_all("table", {"class": "wikitable"})

        if not tablas:
            print("Error: No se encontraron tablas con la clase 'wikitable'.")
            return None

        # Seleccionar la primera tabla (ajustar si es necesario)
        tabla_oscars = tablas[0]

        # Obtener los nombres de las columnas de la tabla
        columnas = [th.getText().strip().lower() for th in tabla_oscars.find_all("th")]

        # Filtrar las columnas de interés
        columnas_deseadas = ['ceremonia', 'mejor película', 'mejor director', 'mejor actor', 'mejor actriz']
        columnas_indices = [columnas.index(col) for col in columnas_deseadas if col in columnas]

        datos = []
        # Recorrer las filas de la tabla
        for fila in tabla_oscars.find_all('tr'):
            # Obtener las celdas de la fila
            celdas = [td.getText().strip() for td in fila.find_all("td")]
            
            # Verificar si hay suficientes celdas en la fila
            if len(celdas) >= len(columnas_indices):
                # Filtrar las celdas según las columnas deseadas
                celdas_filtradas = [celdas[idx] for idx in columnas_indices]
                # Añadir los datos a la lista solo si hay datos válidos y coinciden con las columnas deseadas
                if len(celdas_filtradas) == len(columnas_indices) and any(celdas_filtradas):
                    datos.append(celdas_filtradas)
        
        # Crear el DataFrame con los datos filtrados y las columnas deseadas
        df = pd.DataFrame(datos, columns=columnas_deseadas)
        return df

    else:
        print(f"Error: La petición falló con el código de estado {res_oscars.status_code}")
        return None

# Extraer la tabla de los Oscars y almacenarla en un DataFrame
df_oscars = extraer_tabla_oscars()

# Mostrar las primeras filas del DataFrame resultante
df_oscars
```
